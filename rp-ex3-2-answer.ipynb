# Importando pacotes importantes

```python
import torch
import matplotlib.pyplot as plt
from nn.regression import Regression

from util import fileutils
```

# Coletando base de dados

```python
x, y = fileutils.reader(path='data/dados-ex3-2.csv', splitter=' ')
```

## Calculando os theta's

```python
regression = Regression(x.numpy(), y.numpy(), size=1, ones=False) # gambi
thetas = regression.fit()
theta0 = thetas[0]
ypred = theta0 * x # 
print(f"h2(x) rmse {regression.rmse(ypred, y=y)}")

print("-------------------------------------------")

regression = Regression(x.numpy(), y.numpy(), size=1, ones=True) # gambi
thetas = regression.fit()
theta0, theta1 = thetas
ypred2 = theta1*x + theta0
print(f"h1(x) rmse {regression.rmse(ypred2, y=y)}")
```

# Plotando os graficos

```python
plt.scatter(x, y, c='b', label='csv data')
plt.scatter(x, ypred, c='y', label='h2(x) = \u03B8x')
plt.scatter(x, ypred2, c='g', label='h1(x) = $\u03B8_0$ + $\u03B8_1$x')
plt.title('Base de treinamento')
plt.xlabel('$\u03B8_0$')
plt.ylabel('$\u03B8_1$')
plt.legend()
plt.show()
```

# Conclusão
Baseado no gráfico acima e no resultado do erro quadrático, podemos concluir que o modelo que melhor se aproxima é o modelo h1. Neste modelo, temos o elemento que chamamos de _bias_, este elemento da uma flexibilidade maior à função. No caso do modelo h2, a função é obrigada a passar pela origem, pois não possui o bias.
